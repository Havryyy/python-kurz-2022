```python
import pandas
import pytemperature
```

```python
teplota = pandas.read_csv("temperature.csv")
teplota.head()
```

Teploty jsou ve Fahrenheitech

```python
print(teplota[teplota["City"].isin(["Prague"])])
```

```python
vice_nez_80 = teplota[teplota["AvgTemperature"] > 80 ]
vice_nez_80
```

```python
vice_nez_80_a_v_evrope = teplota[(teplota["AvgTemperature"] > 60) & (teplota["Region"] == "Europe") ]
vice_nez_80_a_v_evrope
```

```python
extremni_hodnoty = teplota[(teplota["AvgTemperature"] > 80) | (teplota["AvgTemperature"] < -20 )]
extremni_hodnoty
```

```python
teplota["AvgTemperatureCelsia"] = pytemperature.f2c(teplota["AvgTemperature"])
teplota
```

```python
vice_nez_15_a_v_evrope = teplota[(teplota["AvgTemperatureCelsia"] > 15) & (teplota["Region"] == "Europe") ]
vice_nez_15_a_v_evrope
```

```python
extremni_hodnoty_Celsius = teplota[(teplota["AvgTemperatureCelsia"] > 30) | (teplota["AvgTemperatureCelsia"] < -10 )]
extremni_hodnoty_Celsius
```

Podezřelé jsou hodnoty průměné teploty "-99"
